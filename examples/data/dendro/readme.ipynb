Weather station data was retrieved from NCEI:
[https://www.ncei.noaa.gov/pub/data/ghcn/daily](https://www.ncei.noaa.gov/pub/data/ghcn/daily)

The weather station data for Hoseda Hard was extracted
and modified such that only data from the All-Russian
Research Institute of Hydrometeorological Information
was included. Any data with a questionable quality flag
was removed completely.

Ring width data is from the International Tree Ring
Database (NOAA): site id = russ208. Reference:
Forbes, B.C., M. Macias Fauria, and P. Zetterberg. 2010.
Russian Arctic warming and 'greening' are closely
tracked by tundra shrub willows.
Global Change Biology 16: 1542-1554.
DOI: 10.1111/j.1365-2486.2009.02047.x.
